## Çoklu Data ve Legend Verme - Plotting multiple sets of data

```python
age = np.arange(25, 44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2 = [45372, 46852, 69401, 56790, 73000, 75000, 80000, 83454]
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b:", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY 2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b:", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY 2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(loc = "lower right") # legend yeri değişti
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b:", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY 2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(loc = 5) # legend yeri değişti
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b:", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY 2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(bbox_to_anchor=(0,0,0,0)) # legend yeri değişti
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b:", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY 2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(bbox_to_anchor=(0,0,1,1)) # legend yeri değişti
plt.show()
```

## 1.3 Figure ve Axes Renklendirmesi

```python
plt.figure(facecolor="yellow", figsize = (10, 5))
ax = plt.axes()
ax.set_facecolor("xkcd:mint green")
plt.plot(age, salary, "b-", linewidth = 4, label = "SALARY 1")
plt.plot(age, salary_2, "r", linestyle  = ":", label = "SALARY 2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title(label = "SALARY BY AGE")
plt.legend(bbox_to_anchor = (0.4, 0.5, 0.5, 0.5))
plt.show()
```

```python
plt.figure(facecolor="crimson", figsize = (10, 5))
ax = plt.axes()
ax.set_facecolor("xkcd:mint green")
plt.plot(age, salary, "b-", lw = 6, label = "SALARY 1")
plt.plot(age, salary_2, "rv", ls = ":", label = "SALARY 2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title(label = "SALARY BY AGE")
plt.legend(bbox_to_anchor = (0.4, 0.5, 0.5, 0.5))
plt.show()
```

```python
plt.figure(facecolor="crimson", figsize = (10, 5))
ax = plt.axes()
ax.set_facecolor("xkcd:mint green")
plt.plot(age, salary, "b-", lw = 6, label = "SALARY 1")
plt.plot(age, salary_2, "ro", ls = "-", label = "SALARY 2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title(label = "SALARY BY AGE")
plt.legend(loc = "upper center")
plt.show()
```

## 1.4. Functional Method

```python
plt.subplot(2, 3, 1)
plt.plot(age, salary, "b")
plt.subplot(2, 3, 2)
plt.plot(age, salary, "b")
plt.subplot(2, 3, 3)
plt.plot(age, salary, "b")
plt.subplot(2, 3, 4)
plt.plot(age, salary_2, "r")
plt.subplot(2, 3, 5)
plt.plot(age, salary_2, "r")
plt.subplot(2, 3, 6)
plt.plot(age, salary_2, "r");

%matplotlib inline
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "b")
plt.tight_layout() #grifiğin birbirine geçmesini engelliyor
plt.show()

fig, ax = plt.subplots(2, 1)
ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "b")
plt.tight_layout() #grifiğin birbirine geçmesini engelliyor
plt.show()
```

```python
fig, ax = plt.subplots(2, 3)
ax[0][0].plot(age, salary, "b")
ax[0][1].plot(age, salary, "b")
ax[0][2].plot(age, salary, "b")
ax[1][0].plot(age, salary_2, "r")
ax[1][1].plot(age, salary_2, "r")
ax[1][2].plot(age, salary_2, "r")
plt.tight_layout() #grifiğin birbirine geçmesini engelliyor
plt.show()
```

```python
fig, ax = plt.subplots(3, 2)
ax[0][0].plot(age, salary, "b")
ax[0][1].plot(age, salary, "b")
ax[1][0].plot(age, salary_2, "r")
ax[1][1].plot(age, salary_2, "r")
ax[2][0].plot(age, salary_2, "r")
ax[2][1].plot(age, salary_2, "r")
plt.tight_layout() #grifiğin birbirine geçmesini engelliyor
plt.show()
```
