## 1 Calling the Libraries

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
```

### Görev 1: Seaborn kütüphanesi içerisinden Titanic veri setini tanımlayınız.

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

### Görev 2: Yukarıda tanımlanan Titanic veri setindeki kadın ve erkek yolcuların sayısını bulunuz.

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='sex', data=titanic)
plt.title('Titanic Veri Setinde Kadın ve Erkek Yolcu Sayısı')

# Çubukların üzerine sayıları yazma
for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

### Görev 3: Her bir sutuna ait unique değerlerin sayısını bulunuz.

```python
unique_counts = titanic.nunique().sort_values(ascending=True) # Her bir sütuna ait unique değerlerin bulunması
plt.figure(figsize=(12, 8)) # Grafik boyutunun ayarlanması 
ax = sns.barplot(x=unique_counts.values, y=unique_counts.index, orient='h') # Yatayda çubuk grafik oluşturulması
plt.title('Titanic Veri Setindeki Her Sütuna Ait Unique Değer Sayısı', fontsize=14) # Başlığın eklenmesi ve boyutlandırılması 

# Eksenlerin etiketle
plt.xlabel('Unique Değer Sayısı', fontsize=12) 
plt.ylabel('Sütunlar', fontsize=12)

# Çubukların üzerine değerlerin yazılması          
for i, v in enumerate(unique_counts.values):
    ax.text(v + 0.1, i, str(v), va='center')
    
#Grafiğin Gösterilmesi
plt.tight_layout()
plt.show()
```

### Görev 4: pclass değişkeninin unique değerleri bulunuz.

```python
pclass_unique_values = titanic['pclass'].unique()
pclass_unique_count_1 = len(pclass_unique_values)

print("unique() kullanarak:")
print(f"pclass'ın benzersiz değerleri: {pclass_unique_values}")
print(f"pclass'ın benzersiz değerlerinin sayısı: {pclass_unique_count_1}")
```

### unique():

##### Bir Series'teki benzersiz değerleri döndürür. Sonuç, benzersiz değerlerin bir listesidir.

### nunique():
##### Bir Series'teki benzersiz değerlerin sayısını döndürür. Sonuç, tek bir sayıdır.  

```python
# pclass değişkeninin unique değerlerinin sayısını bul
pclass_unique_count = titanic['pclass'].nunique()

print(f"pclass değişkeninin unique değerlerinin sayısı: {pclass_unique_count}")

# pclass değerlerinin dağılımını gösteren bir grafik oluşturalım
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='pclass', data=titanic)
plt.title('Yolcu Sınıfı (pclass) Dağılımı')
plt.xlabel('Yolcu Sınıfı')
plt.ylabel('Yolcu Sayısı')

# Çubukların üzerine sayıları yazma
for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

### Görev 5:  pclass ve parch değişkenlerinin unique değerlerinin sayısını bulunuz.

```python
# pclass ve parch değişkenlerinin unique değerlerinin sayısını bul
pclass_unique_count = titanic['pclass'].nunique()
parch_unique_count = titanic['parch'].nunique()

# Unique değer sayılarını gösteren grafik
plt.figure(figsize=(10, 6))
plt.bar(['pclass', 'parch'], [pclass_unique_count, parch_unique_count])
plt.title('pclass ve parch Değişkenlerinin Unique Değer Sayıları')
plt.ylabel('Unique Değer Sayısı')
for i, v in enumerate([pclass_unique_count, parch_unique_count]):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()

# pclass ve parch değerlerinin dağılımını gösteren grafikler
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# pclass dağılımı
sns.countplot(x='pclass', data=titanic, ax=ax1)
ax1.set_title('Yolcu Sınıfı (pclass) Dağılımı')
ax1.set_xlabel('Yolcu Sınıfı')
ax1.set_ylabel('Yolcu Sayısı')
for i in ax1.containers:
    ax1.bar_label(i)

# parch dağılımı
sns.countplot(x='parch', data=titanic, ax=ax2)
ax2.set_title('Ebeveyn/Çocuk Sayısı (parch) Dağılımı')
ax2.set_xlabel('Ebeveyn/Çocuk Sayısı')
ax2.set_ylabel('Yolcu Sayısı')
for i in ax2.containers:
    ax2.bar_label(i)

plt.tight_layout()
plt.show()

print(f"pclass değişkeninin unique değerlerinin sayısı: {pclass_unique_count}")
print(f"parch değişkeninin unique değerlerinin sayısı: {parch_unique_count}")
```

### Görev 6: embarked değişkeninin tipini kontrol ediniz. Tipini category olarak değiştiriniz. Tekrar tipini kontrol ediniz.

```python
print(f"\nGörev 6: embarked değişkeninin orijinal tipi: {titanic['embarked'].dtype}")
titanic['embarked'] = titanic['embarked'].astype('category')
print(f"embarked değişkeninin yeni tipi: {titanic['embarked'].dtype}")
```

### Görev 7: embarked değeri C olanların tüm bilgelerini gösteriniz.

```python

# 'embarked' değeri 'C' olanları filtrele
embarked_c = titanic[titanic['embarked'] == 'C']

# Grafikleri oluştur
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle("Cherbourg'dan (C) Binen Yolcuların Özellikleri", fontsize=16)

# Yaş dağılımı
sns.histplot(data=embarked_c, x='age', kde=True, ax=ax1)
ax1.set_title('Yaş Dağılımı')
ax1.set_xlabel('Yaş')
ax1.set_ylabel('Yolcu Sayısı')

# Cinsiyet dağılımı
sns.countplot(data=embarked_c, x='sex', ax=ax2)
ax2.set_title('Cinsiyet Dağılımı')
ax2.set_xlabel('Cinsiyet')
ax2.set_ylabel('Yolcu Sayısı')
for i in ax2.containers:
    ax2.bar_label(i)

# Yolcu sınıfı dağılımı
sns.countplot(data=embarked_c, x='pclass', ax=ax3)
ax3.set_title('Yolcu Sınıfı Dağılımı')
ax3.set_xlabel('Yolcu Sınıfı')
ax3.set_ylabel('Yolcu Sayısı')
for i in ax3.containers:
    ax3.bar_label(i)

# Hayatta kalma oranı
sns.countplot(data=embarked_c, x='survived', ax=ax4)
ax4.set_title('Hayatta Kalma Oranı')
ax4.set_xlabel('Hayatta Kaldı mı? (0: Hayır, 1: Evet)')
ax4.set_ylabel('Yolcu Sayısı')
for i in ax4.containers:
    ax4.bar_label(i)

plt.tight_layout()
plt.show()

# Bazı istatistiksel bilgileri yazdır
print(f"Cherbourg'dan (C) binen toplam yolcu sayısı: {len(embarked_c)}")
print(f"Ortalama yaş: {embarked_c['age'].mean():.2f}")
print(f"Hayatta kalma oranı: {(embarked_c['survived'].mean()*100):.2f}%")
print("\nYolcu sınıfı dağılımı:")
print(embarked_c['pclass'].value_counts(normalize=True).sort_index())
print("\nCinsiyet dağılımı:")
print(embarked_c['sex'].value_counts(normalize=True))
```

### Görev 8: embarked değeri S olmayanların tüm bilgelerini gösteriniz.

```python

# 'embarked' değeri 'S' olmayanları filtrele
not_s = titanic[titanic['embarked'] != 'S']

# Grafikleri oluştur
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle("Southampton (S) Dışındaki Limanlardan Binen Yolcuların Özellikleri", fontsize=16)

# Yaş dağılımı
sns.histplot(data=not_s, x='age', kde=True, ax=ax1)
ax1.set_title('Yaş Dağılımı')
ax1.set_xlabel('Yaş')
ax1.set_ylabel('Yolcu Sayısı')

# Cinsiyet ve biniş limanı dağılımı
sns.countplot(data=not_s, x='sex', hue='embarked', ax=ax2)
ax2.set_title('Cinsiyet ve Biniş Limanı Dağılımı')
ax2.set_xlabel('Cinsiyet')
ax2.set_ylabel('Yolcu Sayısı')
for container in ax2.containers:
    ax2.bar_label(container)

# Yolcu sınıfı dağılımı
sns.countplot(data=not_s, x='pclass', hue='embarked', ax=ax3)
ax3.set_title('Yolcu Sınıfı Dağılımı')
ax3.set_xlabel('Yolcu Sınıfı')
ax3.set_ylabel('Yolcu Sayısı')
for container in ax3.containers:
    ax3.bar_label(container)

# Hayatta kalma oranı
sns.countplot(data=not_s, x='survived', hue='embarked', ax=ax4)
ax4.set_title('Hayatta Kalma Oranı')
ax4.set_xlabel('Hayatta Kaldı mı? (0: Hayır, 1: Evet)')
ax4.set_ylabel('Yolcu Sayısı')
for container in ax4.containers:
    ax4.bar_label(container)

plt.tight_layout()
plt.show()

# Karşılaştırmalı istatistikler
print("Southampton (S) dışındaki limanlardan binen yolcuların istatistikleri:")
print(f"Toplam yolcu sayısı: {len(not_s)}")
print(f"Ortalama yaş: {not_s['age'].mean():.2f}")
print(f"Hayatta kalma oranı: {(not_s['survived'].mean()*100):.2f}%")
print("\nBiniş limanı dağılımı:")
print(not_s['embarked'].value_counts(normalize=True))
print("\nYolcu sınıfı dağılımı:")
print(not_s['pclass'].value_counts(normalize=True).sort_index())
print("\nCinsiyet dağılımı:")
print(not_s['sex'].value_counts(normalize=True))

print("\n\nSouthampton'dan (S) binen yolcuların istatistikleri:")
s_passengers = titanic[titanic['embarked'] == 'S']
print(f"Toplam yolcu sayısı: {len(s_passengers)}")
print(f"Ortalama yaş: {s_passengers['age'].mean():.2f}")
print(f"Hayatta kalma oranı: {(s_passengers['survived'].mean()*100):.2f}%")
```

### Görev 9: Yaşı 30 dan küçük ve kadın olan yolcuların tüm bilgilerini gösteriniz.

```python

# Yaşı 30'dan küçük ve kadın olan yolcuları filtrele
young_women = titanic[(titanic['age'] < 30) & (titanic['sex'] == 'female')]

# Grafikleri oluştur
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle("30 Yaşından Küçük Kadın Yolcuların Özellikleri", fontsize=16)

# Yaş dağılımı
sns.histplot(data=young_women, x='age', kde=True, ax=ax1)
ax1.set_title('Yaş Dağılımı')
ax1.set_xlabel('Yaş')
ax1.set_ylabel('Yolcu Sayısı')

# Yolcu sınıfı dağılımı
sns.countplot(data=young_women, x='pclass', ax=ax2)
ax2.set_title('Yolcu Sınıfı Dağılımı')
ax2.set_xlabel('Yolcu Sınıfı')
ax2.set_ylabel('Yolcu Sayısı')
for container in ax2.containers:
    ax2.bar_label(container)

# Biniş limanı dağılımı
sns.countplot(data=young_women, x='embarked', ax=ax3)
ax3.set_title('Biniş Limanı Dağılımı')
ax3.set_xlabel('Biniş Limanı')
ax3.set_ylabel('Yolcu Sayısı')
for container in ax3.containers:
    ax3.bar_label(container)

# Hayatta kalma oranı
sns.countplot(data=young_women, x='survived', ax=ax4)
ax4.set_title('Hayatta Kalma Oranı')
ax4.set_xlabel('Hayatta Kaldı mı? (0: Hayır, 1: Evet)')
ax4.set_ylabel('Yolcu Sayısı')
for container in ax4.containers:
    ax4.bar_label(container)

plt.tight_layout()
plt.show()

# Karşılaştırmalı istatistikler
print("30 yaşından küçük kadın yolcuların istatistikleri:")
print(f"Toplam yolcu sayısı: {len(young_women)}")
print(f"Ortalama yaş: {young_women['age'].mean():.2f}")
print(f"Hayatta kalma oranı: {(young_women['survived'].mean()*100):.2f}%")
print("\nYolcu sınıfı dağılımı:")
print(young_women['pclass'].value_counts(normalize=True).sort_index())
print("\nBiniş limanı dağılımı:")
print(young_women['embarked'].value_counts(normalize=True))

print("\n\nDiğer yolcuların istatistikleri:")
other_passengers = titanic[~((titanic['age'] < 30) & (titanic['sex'] == 'female'))]
print(f"Toplam yolcu sayısı: {len(other_passengers)}")
print(f"Ortalama yaş: {other_passengers['age'].mean():.2f}")
print(f"Hayatta kalma oranı: {(other_passengers['survived'].mean()*100):.2f}%")

# Fare (bilet ücreti) karşılaştırması
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='fare', hue='age_group', data=titanic.assign(age_group=lambda x: (x['age'] < 30).map({True: '< 30', False: '≥ 30'})))
plt.title('Cinsiyet ve Yaş Grubuna Göre Bilet Ücreti Dağılımı')
plt.xlabel('Cinsiyet')
plt.ylabel('Bilet Ücreti')
plt.show()
```

### Görev 13: deck değikenindeki boş değerleri deck değişkenin en çok tekrar eden değeri (mode) ile doldurunuz.

```python
# Deck değişkenindeki boş değerlerin sayısını bul
null_count = titanic['deck'].isnull().sum()
print(f"Deck değişkenindeki boş değer sayısı: {null_count}")

# En çok tekrar eden değeri (mode) ve sayısını bul
deck_mode = titanic['deck'].mode()[0]
mode_count = titanic['deck'].value_counts().max()
print(f"En çok tekrar eden değer (mode): {deck_mode}")
print(f"En çok tekrar eden değerin sayısı: {mode_count}")

# Grafikleri oluştur
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Orijinal dağılım
sns.countplot(data=titanic, x='deck', ax=ax1)
ax1.set_title('Orijinal Deck Değişkeninin Dağılımı')
ax1.set_xlabel('Deck')
ax1.set_ylabel('Yolcu Sayısı')
ax1.tick_params(axis='x', rotation=45)
for i in ax1.containers:
    ax1.bar_label(i)

# Boş değerleri doldurulmuş dağılım
titanic['deck_filled'] = titanic['deck'].fillna(deck_mode)
sns.countplot(data=titanic, x='deck_filled', ax=ax2)
ax2.set_title('Deck Değişkeninin Dağılımı (Boş Değerler Doldurulmuş)')
ax2.set_xlabel('Deck')
ax2.set_ylabel('Yolcu Sayısı')
ax2.tick_params(axis='x', rotation=45)
for i in ax2.containers:
    ax2.bar_label(i)

plt.tight_layout()
plt.show()
```

Görev-13 için detaylı açıklama: 

Boş Değerler:
Deck değişkeninde toplam 688 boş değer bulunmaktadır. Bu, Titanic veri setindeki 891 yolcunun büyük bir kısmının (yaklaşık %77'sinin) güverte bilgisinin eksik olduğunu gösteriyor.

En Çok Tekrar Eden Değer (Mode):
Deck değişkeninin en çok tekrar eden değeri 'C' güvertesidir ve bu değer 59 kez tekrar etmektedir.

Grafikler Analizi:
a) Orijinal Dağılım (Sol Grafik):

Veri setinde en çok yolcuya sahip güverte C'dir (59 yolcu). Bunu sırasıyla B (47 yolcu), D (33 yolcu), E (32 yolcu), A (15 yolcu), F (13 yolcu) ve G (4 yolcu) güverteleri takip etmektedir. Grafikte gösterilmeyen büyük bir boşluk var, bu da boş değerleri temsil ediyor.

b) Doldurulmuş Dağılım (Sağ Grafik):

Boş değerler en çok tekrar eden değer olan 'C' ile doldurulduktan sonra, C güvertesindeki yolcu sayısı dramatik bir şekilde 747'ye yükselmiştir (59 + 688).
Diğer güvertelerin yolcu sayıları değişmemiştir.



### Görev 14: age değikenindeki boş değerleri age değişkenin medyanı ile doldurun.

```python

# Yaş değişkenindeki boş değerlerin sayısını bul
null_count = titanic['age'].isnull().sum()
print(f"Yaş değişkenindeki boş değer sayısı: {null_count}")

# Yaş değişkeninin medyanını bul
age_median = titanic['age'].median()
print(f"Yaş değişkeninin medyanı: {age_median:.2f}")

# Grafikleri oluştur
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Orijinal dağılım
sns.histplot(data=titanic, x='age', kde=True, ax=ax1)
ax1.axvline(age_median, color='r', linestyle='--', label=f'Medyan ({age_median:.2f})')
ax1.set_title('Orijinal Yaş Dağılımı')
ax1.set_xlabel('Yaş')
ax1.set_ylabel('Yolcu Sayısı')
ax1.legend()

# Boş değerleri doldurulmuş dağılım
titanic['age_filled'] = titanic['age'].fillna(age_median)
sns.histplot(data=titanic, x='age_filled', kde=True, ax=ax2)
ax2.axvline(age_median, color='r', linestyle='--', label=f'Medyan ({age_median:.2f})')
ax2.set_title('Yaş Dağılımı (Boş Değerler Doldurulmuş)')
ax2.set_xlabel('Yaş')
ax2.set_ylabel('Yolcu Sayısı')
ax2.legend()

plt.tight_layout()
plt.show()

# Doldurma işleminin etkisini göster
print("\nYaş istatistikleri:")
print("Orijinal veri:")
print(titanic['age'].describe())
print("\nDoldurulmuş veri:")
print(titanic['age_filled'].describe())
```

Görev-14 için Detaylı Analiz ve Açıklama:

Yaş Medyanı:
Yaş değişkeninin medyanı 28.0'dir. Bu değer, boş değerleri doldurmak için kullanılacaktır.
Grafikler Analizi:
a) Orijinal Dağılım (Sol Grafik):

- Yaş dağılımı sağa çarpık bir dağılım gösteriyor.
- Çoğu yolcunun 20-40 yaş aralığında olduğu görülüyor.
- Kırmızı kesikli çizgi medyanı gösteriyor (28.0 yaş).
- Grafikte bazı boşluklar var, bu da boş değerleri temsil ediyor.

b) Doldurulmuş Dağılım (Sağ Grafik):

Boş değerler medyan (28.0) ile doldurulduktan sonra, dağılımın genel şekli korunmuş ancak 28 yaş civarında belirgin bir artış gözlemlenmiştir. Medyan değerinde (kırmızı kesikli çizgi) bir tepe noktası oluşmuştur.


İstatistiksel Karşılaştırma:
Orijinal ve doldurulmuş verilerin istatistiklerini karşılaştırdığımızda:

- Ortalama yaş çok az değişmiştir (orijinalde ~29.70, doldurulmuşta ~29.36).
- Minimum ve maksimum yaşlar değişmemiştir.
- Çeyrekler arası aralık (IQR) biraz daralmıştır.
- Standart sapma azalmıştır, bu da dağılımın biraz daha sıkılaştığını gösterir.

### Görev 15: survived değişkeninin Pclass ve Cinsiyet değişkenleri kırılımınında sum, count, mean değerlerini bulunuz.

```python

# survived değişkeninin pclass ve cinsiyet kırılımında istatistiklerini hesapla
survival_stats = titanic.groupby(['pclass', 'sex'])['survived'].agg(['sum', 'count', 'mean'])
survival_stats = survival_stats.reset_index()
survival_stats['death_count'] = survival_stats['count'] - survival_stats['sum']
survival_stats['death_rate'] = 1 - survival_stats['mean']

# Grafikleri oluştur
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 16))

# 1. Hayatta kalma oranı
sns.barplot(x='pclass', y='mean', hue='sex', data=survival_stats, ax=ax1)
ax1.set_title('Yolcu Sınıfı ve Cinsiyete Göre Hayatta Kalma Oranı')
ax1.set_xlabel('Yolcu Sınıfı')
ax1.set_ylabel('Hayatta Kalma Oranı')
for i in ax1.containers:
    ax1.bar_label(i, fmt='%.2f')

# 2. Toplam yolcu sayısı
sns.barplot(x='pclass', y='count', hue='sex', data=survival_stats, ax=ax2)
ax2.set_title('Yolcu Sınıfı ve Cinsiyete Göre Toplam Yolcu Sayısı')
ax2.set_xlabel('Yolcu Sınıfı')
ax2.set_ylabel('Yolcu Sayısı')
for i in ax2.containers:
    ax2.bar_label(i)

# 3. Hayatta kalan ve ölen yolcu sayısı
survival_stats_melted = pd.melt(survival_stats, 
                                id_vars=['pclass', 'sex'], 
                                value_vars=['sum', 'death_count'],
                                var_name='status', value_name='passenger_count')
survival_stats_melted['status'] = survival_stats_melted['status'].map({'sum': 'Survived', 'death_count': 'Died'})

sns.barplot(x='pclass', y='passenger_count', hue='status', data=survival_stats_melted, ax=ax3)
ax3.set_title('Yolcu Sınıfına Göre Hayatta Kalan ve Ölen Yolcu Sayısı')
ax3.set_xlabel('Yolcu Sınıfı')
ax3.set_ylabel('Yolcu Sayısı')
for i in ax3.containers:
    ax3.bar_label(i)

# 4. Cinsiyet ve sınıfa göre hayatta kalma oranı (heatmap)
survival_pivot = survival_stats.pivot(index='sex', columns='pclass', values='mean')
sns.heatmap(survival_pivot, annot=True, fmt='.2%', cmap='YlOrRd', ax=ax4)
ax4.set_title('Cinsiyet ve Yolcu Sınıfına Göre Hayatta Kalma Oranı')
ax4.set_xlabel('Yolcu Sınıfı')
ax4.set_ylabel('Cinsiyet')

plt.tight_layout()
plt.show()

# İstatistikleri yazdır
print(survival_stats)
```

### Görev 16:  30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 vericek bir fonksiyon yazınız.
#### Yazdığınız fonksiyonu kullanarak titanik veri setinde age_flag adında bir değişken oluşturunuz oluşturunuz. (apply ve lambda yapılarını kullanınız)

```python
def create_age_flag(dataframe, age_column='age', threshold=30):
    """
    Verilen bir DataFrame'de yaş sütununa göre bir bayrak (flag) oluşturur.
    
    Parameters:
    -----------
    dataframe : pandas.DataFrame
        İşlem yapılacak DataFrame.
    age_column : str, optional (default='age')
        Yaş bilgisini içeren sütunun adı.
    threshold : int, optional (default=30)
        Yaş eşik değeri.
    
    Returns:
    --------
    pandas.DataFrame
        'age_flag' sütunu eklenmiş yeni DataFrame.
    
    Notes:
    ------
    - Fonksiyon, yaşı threshold değerinden küçük olanlara 1, 
      eşit ve büyük olanlara 0 değerini atar.
    - NaN değerler korunur.
    - Orijinal DataFrame değiştirilmez, yeni bir kopya döndürülür.
    """
    
    # DataFrame'in bir kopyasını oluştur
    df = dataframe.copy()
    
    # age_flag sütununu oluştur
    df['age_flag'] = df[age_column].apply(lambda x: 1 if x < threshold else 0 if pd.notnull(x) else x)
    
    return df

# Fonksiyonu test edelim
titanic = sns.load_dataset('titanic')
titanic_with_flag = create_age_flag(titanic)

# Sonuçları kontrol edelim
print(titanic_with_flag[['age', 'age_flag']].head(10))
print("\nYaş bayraklarının dağılımı:")
print(titanic_with_flag['age_flag'].value_counts(dropna=False))
print("\nOrijinal yaş dağılımı:")
print(titanic_with_flag['age'].describe())
```

### Görev 18: Time değişkeninin kategorilerine (Dinner, Lunch) göre total_bill  değerlerinin toplamını, min, max ve ortalamasını bulunuz.

```python
# Time değişkenine göre total_bill değerlerinin istatistiklerini hesapla
# observed=True parametresini ekleyerek uyarıyı giderelim
time_stats = tips.groupby('time', observed=True)['total_bill'].agg(['sum', 'min', 'max', 'mean'])

# Sonuçları yazdır
print("Time değişkenine göre total_bill istatistikleri:")
print(time_stats)
print("\nDetaylı açıklama:")
for time, stats in time_stats.iterrows():
    print(f"\n{time.capitalize()}:")
    print(f"  Toplam hesap: ${stats['sum']:.2f}")
    print(f"  En düşük hesap: ${stats['min']:.2f}")
    print(f"  En yüksek hesap: ${stats['max']:.2f}")
    print(f"  Ortalama hesap: ${stats['mean']:.2f}")

# Grafikleri oluştur
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Time Değişkenine Göre Total Bill Analizi', fontsize=16)

# 1. Toplam hesap
sns.barplot(x=time_stats.index, y='sum', data=time_stats, ax=ax1)
ax1.set_title('Toplam Hesap')
ax1.set_ylabel('Toplam ($)')
for i, v in enumerate(time_stats['sum']):
    ax1.text(i, v, f'${v:.2f}', ha='center', va='bottom')

# 2. Ortalama hesap
sns.barplot(x=time_stats.index, y='mean', data=time_stats, ax=ax2)
ax2.set_title('Ortalama Hesap')
ax2.set_ylabel('Ortalama ($)')
for i, v in enumerate(time_stats['mean']):
    ax2.text(i, v, f'${v:.2f}', ha='center', va='bottom')

# 3. Minimum ve maksimum hesap
time_stats[['min', 'max']].plot(kind='bar', ax=ax3)
ax3.set_title('Minimum ve Maksimum Hesap')
ax3.set_ylabel('Tutar ($)')
ax3.legend(['Minimum', 'Maksimum'])
for i, row in enumerate(time_stats[['min', 'max']].itertuples()):
    ax3.text(i, row.min, f'${row.min:.2f}', ha='center', va='bottom')
    ax3.text(i, row.max, f'${row.max:.2f}', ha='center', va='bottom')

# 4. Hesap dağılımı (box plot)
sns.boxplot(x='time', y='total_bill', data=tips, ax=ax4)
ax4.set_title('Hesap Dağılımı')
ax4.set_ylabel('Hesap ($)')

plt.tight_layout()
plt.show()
```

### Görev 21: size'i 3'ten küçük, total_bill'i 10'dan büyük olan siparişlerin ortalaması nedir?

```python

# Filtreleme işlemi (loc kullanarak)
filtered_tips = tips.loc[(tips['size'] < 3) & (tips['total_bill'] > 10)]

# Ortalama hesaplama
average_bill = filtered_tips['total_bill'].mean()

print(f"Size'i 3'ten küçük ve total_bill'i 10'dan büyük olan siparişlerin ortalama total_bill değeri: ${average_bill:.2f}")

# Görselleştirme
plt.figure(figsize=(12, 6))

# Tüm veri seti için scatter plot
sns.scatterplot(data=tips, x='size', y='total_bill', alpha=0.5, label='Tüm Siparişler')

# Filtrelenmiş veri için scatter plot
sns.scatterplot(data=filtered_tips, x='size', y='total_bill', color='red', label='Filtrelenmiş Siparişler')

plt.axhline(y=10, color='green', linestyle='--', label='Total Bill = $10')
plt.axvline(x=3, color='purple', linestyle='--', label='Size = 3')
plt.axhline(y=average_bill, color='red', linestyle='--', label=f'Ortalama = ${average_bill:.2f}')

plt.title("Size ve Total Bill İlişkisi")
plt.xlabel("Grup Büyüklüğü")
plt.ylabel("Toplam Hesap ($)")
plt.legend()

plt.tight_layout()
plt.show()

# Ek istatistikler
print("\nEk İstatistikler:")
print(f"Filtrelenmiş sipariş sayısı: {len(filtered_tips)}")
print(f"Tüm siparişlerin ortalama total_bill değeri: ${tips['total_bill'].mean():.2f}")
print("\nFiltrelenmiş siparişlerin özet istatistikleri:")
print(filtered_tips['total_bill'].describe())

# Filtrelenmiş siparişlerin dağılımı
plt.figure(figsize=(10, 6))
sns.histplot(data=filtered_tips, x='total_bill', kde=True)
plt.axvline(x=average_bill, color='red', linestyle='--', label=f'Ortalama = ${average_bill:.2f}')
plt.title("Filtrelenmiş Siparişlerin Total Bill Dağılımı")
plt.xlabel("Toplam Hesap ($)")
plt.ylabel("Frekans")
plt.legend()
plt.show()
```
