```python
def data_preparation(df):
    """
    Veriyi hazırlama sürecini fonksiyon haline getiren fonksiyon.
    
    Args:
    df (DataFrame): İşlenecek veri seti.
    
    Returns:
    df (DataFrame): Hazırlanan ve işlenmiş veri seti.
    """
    # 1. Tarih değişkenlerini datetime formatına çevirme
    date_columns = ["first_order_date", "last_order_date", "last_order_date_online", "last_order_date_offline"]
    df[date_columns] = df[date_columns].apply(pd.to_datetime)
    
    # 2. Toplam alışveriş sayısı ve toplam harcama değişkenlerini oluşturma
    df["total_order_num"] = df["order_num_total_ever_online"] + df["order_num_total_ever_offline"]
    df["total_customer_value"] = df["customer_value_total_ever_online"] + df["customer_value_total_ever_offline"]
    
    # 3. Eksik değer kontrolü ve gerekiyorsa düzenleme
    # Bu veri setinde eksik veri işlemi gerekmeyebilir ama genelde bir adım olarak eklenebilir.
    # df.fillna(0, inplace=True) # eksik değerleri 0 ile doldurma örneği
    
    return df

# Fonksiyonu uygulayarak veriyi tekrar hazırlayalım
prepared_data = data_preparation(data_copy)

# İlk 5 gözlem ile hazırlanan veri setini gözden geçirelim
prepared_data.head()
```

## Görev 2: RFM Metriklerinin Hesaplanması 

### Adım 1: Recency, Frequency ve Monetary tanımlarını yapınız.

### 1. **Recency (Güncellik)**:
   - **Tanım**: Müşterinin en son alışveriş yaptığı tarihten bugüne kadar geçen süreyi ifade eder. Recency, bir müşterinin ne kadar "güncel" olduğunu gösterir. Pazarlamada, yakın zamanda alışveriş yapmış müşterilerin yeniden alışveriş yapma olasılığı daha yüksektir.
   - **Nasıl Hesaplanır?**: "Bugünün tarihi" ile müşterinin **son alışveriş tarihi** arasındaki fark gün olarak hesaplanır.
   - **Örnek**: Eğer bir müşteri 30 gün önce alışveriş yapmışsa, Recency değeri 30 olacaktır.

### 2. **Frequency (Sıklık)**:
   - **Tanım**: Müşterinin belirli bir zaman diliminde kaç kez alışveriş yaptığını ifade eder. Frequency, müşterinin şirkete ne kadar sadık olduğunu gösterir. Sıklık değeri yüksek olan müşteriler, daha sık alışveriş yapan ve markaya daha bağlı olan müşterilerdir.
   - **Nasıl Hesaplanır?**: Müşterinin **toplam alışveriş sayısı** ile hesaplanır. Bu veri genellikle online ve offline alışverişlerin toplamı ile elde edilir.
   - **Örnek**: Eğer bir müşteri belirli bir dönemde 10 kez alışveriş yapmışsa, Frequency değeri 10 olacaktır.

### 3. **Monetary (Para Değeri)**:
   - **Tanım**: Müşterinin yaptığı alışverişlerin toplam parasal değerini ifade eder. Monetary, müşterinin şirkete ne kadar kazanç sağladığını gösterir. Daha yüksek Monetary değeri, müşterinin daha fazla harcama yaptığını ve dolayısıyla şirket için daha değerli olduğunu gösterir.
   - **Nasıl Hesaplanır?**: Müşterinin **toplam harcaması** (online ve offline alışverişlerin toplamı) ile hesaplanır.
   - **Örnek**: Eğer bir müşteri toplamda 1.000 TL harcamışsa, Monetary değeri 1.000 olacaktır.

### Adım 2: Müşteri özelinde Recency, Frequency ve Monetary metriklerini hesaplayınız.

```python
import datetime as dt
```

```python
# Veri setindeki maksimum tarih
max_date = data_copy['last_order_date'].max()
```

```python
# Analiz tarihini maksimum tarihten 2 gün sonrası olarak belirleme
today_date = max_date + dt.timedelta(days=2)
```

```python
# Adım 2: Recency, Frequency ve Monetary hesaplama
data_copy['recency'] = (today_date - data_copy['last_order_date']).dt.days
data_copy['frequency'] = data_copy['total_order_num']
data_copy['monetary'] = data_copy['total_customer_value']
```

```python
# 3. Monetary (Para Değeri): Müşterinin toplam harcaması (online ve offline harcamaların toplamı)
data_copy['monetary'] = data_copy['total_customer_value']
```

### Adım 3: Hesapladığınız metrikleri rfm isimli bir değişkene atayınız.

```python
rfm = data_copy[['master_id', 'recency', 'frequency', 'monetary']]
```

### Adım 4: Oluşturduğunuz metriklerin isimlerini recency, frequency ve monetary olarak değiştiriniz.

```python
rfm.columns = ['master_id', 'recency', 'frequency', 'monetary']
rfm.head()
```

## Görev 3: RF Skorunun Hesaplanması

### Adım 1: Recency, Frequency ve Monetary metriklerini qcut yardımı ile 1-5 arasında skorlara çeviriniz.

```python
recency_score = pd.qcut(rfm['recency'], 5, labels=[5, 4, 3, 2, 1])
frequency_score = pd.qcut(rfm['frequency'].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
monetary_score = pd.qcut(rfm['monetary'], 5, labels=[1, 2, 3, 4, 5])
```

### Adım 2: Bu skorları recency_score, frequency_score ve monetary_score olarak kaydediniz.

```python
rfm.loc[:, 'recency_score'] = recency_score
rfm.loc[:, 'frequency_score'] = frequency_score
rfm.loc[:, 'monetary_score'] = monetary_score
rfm[['recency_score', 'frequency_score', 'monetary_score']].head()
```

```python
rfm.loc[:, 'RF_SCORE'] = rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str)
rfm[['recency_score', 'frequency_score', 'RF_SCORE']].head()
```

## Görev 4: RF Skorunun Segment Olarak Tanımlanması

### Adım 1: Oluşturulan RF skorları için segment tanımlamaları yapınız.

```python
# Segment tanımlamalarını içeren bir sözlük oluşturma
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}
```

### Adım 2: Aşağıdaki seg_map yardımı ile skorları segmentlere çeviriniz.

```python
# RF_SCORE'u seg_map yardımıyla segmentlere çevirme
rfm['segment'] = rfm['RF_SCORE'].replace(seg_map, regex=True)
```

```python
# Sonuçları kontrol etme
rfm[['RF_SCORE', 'segment']].head()
```

## 5: Aksiyon Zamanı!

### Adım1: Segmentlerin recency, frequnecy ve monetary ortalamalarını inceleyiniz.

```python
segment_analysis = rfm.groupby("segment").agg({
    "recency": "mean",
    "frequency": "mean",
    "monetary": "mean"
})
segment_analysis
```
