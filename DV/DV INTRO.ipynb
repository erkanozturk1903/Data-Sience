### Histogram

### Box Plot

```python
x = np.random.normal(170, 10, 250) # 170 mean, 10 SD (std.sap.), 250 gözlem
```

### Pie Chart

```python
x = ['Kategori 1', 'Kategori 2', 'Kategori 3', 'Kategori 4']
# Veri
sizes = [15, 30, 45, 10]

# Pasta grafiğini oluştur
plt.figure(figsize=(10, 10))
plt.pie(sizes, labels=x)

# Göster
plt.axis('equal')
plt.show()
```
