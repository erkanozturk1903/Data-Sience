```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
#Balon boyutu ve fiyat
x = np.array([1,2,3,4]).reshape(-1,1) # Bağımsız değişken (balon boyutu)
y = np.array([2,6,12,20]) # Bağımlı değişken(fiyat)
```

```python
# Doğrusal Regresyon modeli
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
# Polinom regresyon modeli(2. dereceden)
poly_features = PolynomialFeatures(degree=2) # Polinom derecesini buradan değiştirebiliriz
x_poly = poly_features.fit_transform(x)
poly_reg = LinearRegression()
poly_reg.fit(x_poly,y)
```

```python
#Tahminler
x_new = np.linspace(1,4,100).reshape(-1,1) # Daha fazla nokta için
x_new_poly = poly_features.transform(x_new)
y_pred_linear = lin_reg.predict(x_new) # Doğrusal Tahminler
y_pred_poly = poly_reg.predict(x_new_poly) # Polinom Tahminler
```

```python
mse_linear = mean_squared_error(y,lin_reg.predict(x))
mse_poly = mean_squared_error(y,poly_reg.predict(x_poly))
```

```python
#Grafik
plt.figure(figsize=(10,6))
plt.scatter(x,y, color = 'red', label = 'Gerçek Veriler')
plt.plot(x_new, y_pred_linear, color = 'blue', label = f'Doğrusal MSE: {mse_linear:.2f}')
plt.plot(x_new, y_pred_poly, color = 'green', label = f'Polinom MSE: {mse_poly:.2f}')
plt.xlabel('Balon Boyutu')
plt.ylabel('Fiyat')
plt.title('Doğrusal ve Polinom Regresyon Modelleri')
plt.legend()
plt.show()
```
