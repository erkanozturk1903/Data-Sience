# Dedektif Delimiter ve Kayıp Sütunlar Vakası 🕵️‍♂️📊

Bir gün, Veri Şehri'nde, Dedektif Delimiter'a acil bir çağrı geldi. Winequality-red.csv adlı dosya, sütunlarını kaybetmişti! İşte vakamızın detayları:

## Olay Yeri İncelemesi: "delimiter" Parametresi

1. **Görev Tanımı**: 
   Dedektif Delimiter'ın görevi, CSV dosyasındaki kayıp sütunları bulmak ve onları birbirinden ayırmak.

2. **Neden Bu Kadar Önemli?**:
   CSV dosyaları genelde virgülle ayrılmış değerler içerir, ama bazen... Bazen işler karışır! Farklı ayırıcılar kullanılmış olabilir ve Dedektif Delimiter'ın bunu çözmesi gerekir.

3. **Winequality-red.csv Vakası**:
   Dedektif, dosyayı incelediğinde şok edici gerçeği fark etti: Standart virgül yerine noktalı virgül (;) kullanılmıştı! "Aha!" diye bağırdı, "İşte suçlu bu!"

4. **Felaket Senaryosu**:
   Eğer Dedektif Delimiter olaya müdahale etmeseydi:
   - Tüm veriler tek bir sütunda sıkışıp kalabilirdi! 😱
   - Veya daha da kötüsü, veriler yanlış yerlere bölünebilir, sütunlar birbirine karışabilirdi! 🙃
   - Sonuç mu? Tam bir veri kaosu! 🌪️

5. **Dedektif Delimiter'ın Sihirli Dokunuşu**:
   Hemen `delimiter=';'` kodunu yazdı. Ve işte mucize! Tüm sütunlar yerli yerine oturdu.

6. **Diğer Gizli Ayırıcılar**:
   Dedektif Delimiter, diğer vakalarda karşılaşabileceği gizli ayırıcıları da not etti:
   - Tab karakteri: `\t` (Görünmez ama etkili!)
   - Boşluk: `' '` (Bazen en basit şeyler gözden kaçar)

## Sonuç

Dedektif Delimiter, bu vakayı başarıyla çözdü. Winequality-red.csv dosyası artık mükemmel bir şekilde okunabiliyor ve tüm sütunlar mutlu mesut yerlerinde duruyorlar.

Unutmayın, veri dünyasında her zaman uyanık olun! Delimiter'ınızı kontrol etmeden asla bir CSV dosyası açmayın. Yoksa siz de Kayıp Sütunlar Vakası'nın bir kurbanı olabilirsiniz! 🕵️‍♂️🍷📊

### Adım 4: Veri Temizleme ve Ön İşleme

````python
# Aykırı değerleri kontrol etme ve temizleme
Q1 = wine_data.quantile(0.25)
Q3 = wine_data.quantile(0.75)
IQR = Q3 - Q1
wine_data_clean = wine_data[~((wine_data < (Q1 - 1.5 * IQR)) | (wine_data > (Q3 + 1.5 * IQR))).any(axis=1)]

print(f"Orijinal veri seti boyutu: {wine_data.shape}")
print(f"Temizlenmiş veri seti boyutu: {wine_data_clean.shape}")
````
