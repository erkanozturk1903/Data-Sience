```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

Bu satırlar, projemiz için gerekli olan Python kütüphanelerini içe aktarır:

pandas: Veri manipülasyonu ve analizi için kullanılan güçlü bir kütüphane.
train_test_split: Scikit-learn kütüphanesinden, veri setimizi eğitim ve test setlerine bölmek için kullanacağımız fonksiyon.
StandardScaler: Yine Scikit-learn'den, özelliklerimizi ölçeklendirmek için kullanacağımız sınıf.

```python
# Load the Iris dataset
iris_df = pd.read_csv('iris.csv')
```

Bu satır, 'iris.csv' dosyasını okur ve içeriğini bir pandas DataFrame'ine yükler. DataFrame, verileri tablo formatında tutmamızı sağlayan bir veri yapısıdır.

```python
# Separate features and target
X = iris_df.drop('target', axis=1)
y = iris_df['target']
```

Burada veri setimizi iki parçaya ayırıyoruz:

X: Bağımsız değişkenler (özellikler). 'target' sütunu hariç tüm sütunları içerir.
y: Bağımlı değişken (hedef). Sadece 'target' sütununu içerir.

drop fonksiyonu, 'target' sütununu DataFrame'den çıkarır ve geriye kalan tüm sütunları X'e atar.

```python
# Split the dataset into an 80-20 training-test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

Bu satır, veri setimizi eğitim ve test setlerine böler:

test_size=0.2: Veri setinin %20'si test seti, %80'i eğitim seti olacak şekilde bölünür.
random_state=42: Bu, bölme işleminin tekrarlanabilir olmasını sağlar. Her çalıştırmada aynı bölünmeyi elde ederiz.

```python
# Print the original training and test sets
print("Original Training Set:")
print(X_train)
print("\nOriginal Test Set:")
print(X_test)
```

Bu satırlar, bölünmüş orijinal veri setlerini ekrana yazdırır. Bu, ölçeklendirme öncesi verilerin nasıl göründüğünü anlamamızı sağlar.

```python
# Apply feature scaling on the training and test sets
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Bu adımda, StandardScaler kullanarak özelliklerimizi ölçeklendiriyoruz:

scaler.fit_transform(X_train): Bu metod, eğitim setini kullanarak ölçeklendirme parametrelerini hesaplar ve ardından bu parametreleri kullanarak veriyi dönüştürür.
scaler.transform(X_test): Bu metod, eğitim setinden elde edilen parametreleri kullanarak test setini dönüştürür. Test seti üzerinde fit işlemi yapmıyoruz çünkü bu, test verisiyle ilgili bilgiyi modele sızdırmak anlamına gelir.

```python
# Print the scaled training and test sets
print("\nScaled Training Set:")
print(X_train_scaled)
print("\nScaled Test Set:")
print(X_test_scaled)
```

Son olarak, ölçeklendirilmiş eğitim ve test setlerini ekrana yazdırıyoruz. Bu, ölçeklendirme işleminin etkisini görmemizi sağlar.
Bu kod, veri ön işleme sürecinin temel adımlarını gösterir: veri yükleme, bölme ve ölçeklendirme. Bu adımlar, makine öğrenimi modellerini eğitmeden önce yapılan tipik hazırlık aşamalarıdır ve modelin performansını önemli ölçüde etkileyebilir.
